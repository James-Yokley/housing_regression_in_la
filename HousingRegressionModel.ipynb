#Housing Prices Prediction Model


- Imports

```python
import pandas as pd
import numpy as np
from sklearn import metrics
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import scale
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import matplotlib

```

```python
#from google.colab import files
#uploaded = files.upload()
```

```python
df = pd.read_csv('/LA_housing_set.csv')

display(df)


feature_cols = ['zipcode', 'year', 'day', 'month']
zipcodes = [90732, 90272, 91040, 90068, 91602, 91345, 91436, 90293, 90077, 90058, 90011, 91331, 90044, 91342, 91335, 90003]

pred_df = pd.read_csv('/LA_prediction_set.csv')

display(pred_df)

rmse_df = pd.DataFrame({'zipcodes' : zipcodes})

display(rmse_df)
```

#Train Models Based on Zipcode

- Take only the rows that pertain to each zipcode from the list
- Split into train/test
- Fit the Linear Regression model 
- Calculate the RMSE
- Plot the regression line

```python
df = df.dropna()

X = df[feature_cols]
y = df['medianprice']

display(X)
display(y)

my_linreg = LinearRegression()

rmse_list = [] 

for zip in zipcodes: 
  new_df = df[(df.zipcode == zip)]
  X = new_df[feature_cols]
  y = new_df['medianprice']
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
  my_linreg.fit(X_train, y_train)

  print(f'Zipcode: {zip}, Intercept: {my_linreg.intercept_}')
  print(f'Zipcode: {zip}, Coefficient: {my_linreg.coef_}')

  y_prediction = my_linreg.predict(X_test)
  mse = metrics.mean_squared_error(y_test, y_prediction)
  rmse = np.sqrt(mse)
  rmse_list.append(rmse)
  print(f'Zipcode: {zip}, RMSE: {rmse}')
  plt.clf()
  plt.scatter(X_train['year'], y_train, color='green', s=40, alpha=0.5)

  plt.plot(np.sort(X_test['year']), np.sort(y_prediction), color='red')
  plt.title(f'Housing Prices vs Zip Code: {zip}')
  plt.xlim([2000, 2022])
  plt.xlabel("Year")
  plt.ylabel("Housing Price")
  plt.show()
  

  sub_pred_df = pred_df[(pred_df.zipcode == zip)] 
  prediction = my_linreg.predict(sub_pred_df)

  plt.clf()

  plt.plot(np.sort(sub_pred_df['year']), np.sort(prediction), color='red')
  plt.title(f'Housing Prices vs Zip Code: {zip}')
  plt.xlabel("Year")
  plt.ylabel("Housing Price")
  plt.show()

rmse_df.insert(1, 'Linear Regression', rmse_list)
```

# Same Process, with Scaling

- Loop through the zipcodes
- Scale the dataset
- Split into train/test
- Fit the Linear Regression model 
- Calculate the RMSE
- Plot the regression line

```python
rmse_list = [] 

for zip in zipcodes: 
  new_df = df[(df.zipcode == zip)]
  X = new_df[feature_cols]
  y = new_df['medianprice']
  X_scaled = scale(X)
  X = pd.DataFrame(X_scaled, columns=feature_cols)
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
  my_linreg.fit(X_train, y_train)

  print(f'Zipcode: {zip}, Intercept: {my_linreg.intercept_}')
  print(f'Zipcode: {zip}, Coefficient: {my_linreg.coef_}')

  y_prediction = my_linreg.predict(X_test)
  mse = metrics.mean_squared_error(y_test, y_prediction)
  rmse = np.sqrt(mse)
  rmse_list.append(rmse)
  print(f'Zipcode: {zip}, RMSE: {rmse}')
  plt.clf()
  plt.scatter(X_train['year'], y_train, color='green', s=40, alpha=0.5)

  plt.plot(np.sort(X_test['year']), np.sort(y_prediction), color='red')
  plt.title(f'Housing Prices vs Zip Code: {zip}')
  plt.xlabel("Year 2000-2022")
  plt.ylabel("Housing Price")
  plt.show()

  sub_pred_df = pred_df[(pred_df.zipcode == zip)]
  pred_scaled = scale(sub_pred_df)
  sub_pred_df = pd.DataFrame(pred_scaled, columns=feature_cols)

  prediction = my_linreg.predict(sub_pred_df)

  plt.clf()

  plt.plot(np.sort(sub_pred_df['year']), np.sort(prediction), color='red')
  plt.title(f'Housing Prices vs Zip Code: {zip}')
  plt.xlabel("Year 2023-2033")
  plt.ylabel("Housing Price")
  plt.show()

rmse_df.insert(2, 'Linear Regression Scaled', rmse_list)
```

## Decision Tree Regression


```python
from sklearn.tree import DecisionTreeRegressor

X = df[feature_cols]
y = df['medianprice']

rmse_list_five = [] 
rmse_list_seven = [] 
for zip in zipcodes: 
  new_df = df[(df.zipcode == zip)]
  X = new_df[feature_cols]
  y = new_df['medianprice']
  X_scaled = scale(X)
  X = pd.DataFrame(X_scaled, columns=feature_cols)
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)


  regr_depth_5 = DecisionTreeRegressor(max_depth=5)
  regr_depth_7 = DecisionTreeRegressor(max_depth=7)
  regr_depth_5.fit(X_train, y_train)
  regr_depth_7.fit(X_train, y_train)


  depth_5_pred = regr_depth_5.predict(X_test)
  depth_7_pred = regr_depth_7.predict(X_test)

  print(f'Results for {zip}: ')

  result_df = pd.DataFrame({'Real Values' : y_test,  
                            'Predicted Values Depth 5' : depth_5_pred,
                            'Predicted Values Depth 7' : depth_7_pred})
  five_rmse = np.sqrt(metrics.mean_squared_error(y_test, depth_5_pred))
  seven_rmse = np.sqrt(metrics.mean_squared_error(y_test, depth_7_pred))
  rmse_list_five.append(five_rmse)
  rmse_list_seven.append(seven_rmse)
  print(f'Depth 5 RMSE: {five_rmse}')
  print(f'Depth 7 RMSE: {seven_rmse}')

  display(result_df)

  plt.scatter(X_test['year'], y_test, color='red')
  plt.plot(np.sort(X_test['year']), np.sort(y_test), color='red')
  plt.scatter(X_test['year'], depth_5_pred, color='orange')
  plt.scatter(X_test['year'], depth_7_pred, color='blue')
  plt.plot(np.sort(X_test['year']), np.sort(depth_7_pred), color='blue')
  plt.title(f'Housing Decision Tree Regression: {zip}')
  plt.xlabel('Year')
  plt.ylabel('Median Price')
  plt.show()

rmse_df.insert(3, 'Decision Tree Depth 5', rmse_list_five)
rmse_df.insert(4, 'Decision Tree Depth 7', rmse_list_seven)
```

## Random Forest Regression

```python
from sklearn.ensemble import RandomForestRegressor



X = df[feature_cols]
y = df['medianprice']
rmse_list = [] 

for zip in zipcodes: 
  new_df = df[(df.zipcode == zip)]
  X = new_df[feature_cols]
  y = new_df['medianprice']
  X_scaled = scale(X)
  X = pd.DataFrame(X_scaled, columns=feature_cols)
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

  my_forest = RandomForestRegressor(n_estimators=20, max_depth=10, criterion='squared_error')

  forest_regr = my_forest.fit(X_train, y_train)

  forest_pred = my_forest.predict(X_test)

  results_df = pd.DataFrame({'Real Values' : y_test, 
                             'Predicted Values' : forest_pred})
  
  display(results_df)


  print(f'Results for {zip}: ')

  forest_rmse = np.sqrt(metrics.mean_squared_error(y_test, forest_pred))
  rmse_list.append(forest_rmse)
  print(f'RMSE: {forest_rmse}')


  plt.scatter(X_test['year'], y_test, color='red')
  plt.plot(np.sort(X_test['year']), np.sort(y_test), color='red')
  plt.scatter(X_test['year'], forest_pred, color='blue')
  plt.plot(np.sort(X_test['year']), np.sort(forest_pred), color='blue')
  plt.title(f'Housing Decision Tree Regression: {zip}')
  plt.xlabel('Year')
  plt.ylabel('Median Price')
  plt.show()

rmse_df.insert(5, 'Random Forest', rmse_list)
```

## KNN Regression


```python
from sklearn import neighbors



new_df = df[(df.zipcode == 90077)]


X = new_df[feature_cols]
y = new_df['medianprice']
rmse_vals = []
for i in range(50): 
  i = i+1

  my_knn = neighbors.KNeighborsRegressor(n_neighbors=i)
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
  my_knn.fit(X_train, y_train)
  knn_pred = my_knn.predict(X_test)

  error = np.sqrt(metrics.mean_squared_error(y_test, knn_pred))

  rmse_vals.append(error)

  print(f'RMSE for {i}: {error}')


curve = pd.DataFrame(rmse_vals)

curve.plot()
```

- From this, we can see that at k=7, we get the lowest RMSE

- We will use this value to prepare our models

```python
X = df[feature_cols]
y = df['medianprice']

rmse_list = [] 

for zip in zipcodes: 
  new_df = df[(df.zipcode == zip)]
  X = new_df[feature_cols]
  y = new_df['medianprice']
  X_scaled = scale(X)
  X = pd.DataFrame(X_scaled, columns=feature_cols)
  X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)

  my_knn = neighbors.KNeighborsRegressor(n_neighbors=7)
  my_knn.fit(X_train, y_train)
  knn_pred = my_knn.predict(X_test)

  results_df = pd.DataFrame({'Real Values' : y_test, 
                             'Predicted Values' : knn_pred})
  
  display(results_df)


  print(f'Results for {zip}: ')

  knn_rmse = np.sqrt(metrics.mean_squared_error(y_test, knn_pred))
  rmse_list.append(knn_rmse)
  print(f'RMSE: {knn_rmse}')


  plt.scatter(X_test['year'], y_test, color='red')
  plt.plot(np.sort(X_test['year']), np.sort(y_test), color='red')
  plt.scatter(X_test['year'], knn_pred, color='blue')
  plt.plot(np.sort(X_test['year']), np.sort(knn_pred), color='blue')
  plt.title(f'Housing Decision Tree Regression: {zip}')
  plt.xlabel('Year')
  plt.ylabel('Median Price')
  plt.show()

  sub_pred_df = pred_df[(pred_df.zipcode == zip)] 
  sub_scaled = scale(sub_pred_df)
  sub_pred_df = pd.DataFrame(sub_scaled, columns=feature_cols)
  prediction = my_knn.predict(sub_pred_df)

  plt.clf()

  plt.plot(np.sort(sub_pred_df['year']), np.sort(prediction), color='red')
  plt.title(f'Housing Prices vs Zip Code: {zip}')
  plt.xlabel("Year")
  plt.ylabel("Housing Price")
  plt.show()

rmse_df.insert(6, 'KNN', rmse_list)  
```
